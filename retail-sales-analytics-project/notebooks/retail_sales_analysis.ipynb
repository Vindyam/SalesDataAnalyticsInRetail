# Retail Sales Data Analysis

Portfolio-ready analysis of retail transactions using Python and Matplotlib.

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/retail_sales_dataset.csv')
df.head()
```

```
df.shape, df.columns.tolist()
```

```
df.info()
```

```
df.describe(include='all')
```

## Missing values

```
df.isna().sum()
```

## Total sales

```
total_col = "Total Amount"
if total_col:
    print('Total Sales:', df[total_col].sum())
else:
    print('No total sales column was detected automatically.')
```

## Sales by category

```
category_col = "Product Category"
total_col = "Total Amount"
category_sales = df.groupby(category_col)[total_col].sum().sort_values(ascending=False)
category_sales
```

```
plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')
plt.title('Sales by Product Category')
plt.xlabel(category_col)
plt.ylabel(total_col)
plt.tight_layout()
plt.show()
```

## Sales by gender

```
gender_col = "Gender"
gender_sales = df.groupby(gender_col)[total_col].sum().sort_values(ascending=False)
gender_sales
```

```
plt.figure(figsize=(7,5))
gender_sales.plot(kind='bar')
plt.title('Sales by Gender')
plt.xlabel(gender_col)
plt.ylabel(total_col)
plt.tight_layout()
plt.show()
```

## Age distribution

```
age_col = "Age"
plt.figure(figsize=(8,5))
pd.to_numeric(df[age_col], errors='coerce').dropna().plot(kind='hist', bins=20)
plt.title('Customer Age Distribution')
plt.xlabel(age_col)
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Monthly sales trend

```
date_col = "Date"
df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
monthly_sales = (
    df.dropna(subset=[date_col])
      .assign(month=df[date_col].dt.to_period('M').astype(str))
      .groupby('month')[total_col]
      .sum()
)
monthly_sales
```

```
plt.figure(figsize=(9,5))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel(total_col)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
